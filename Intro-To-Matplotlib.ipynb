 ## Matplotlib Example Workflow

```python
# 1. import matplotlib and get ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 2. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 3. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width , height)
# 4. plot data
ax.plot(x,y)

# 5. Customize plot
ax.set(title='simple plot', xlabel='x-axis',ylabel='y-axis')

# 6. save and show figure
fig.savefig('./images/sample-plot.png')
```

 ### Two option for subplots

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                              ncols=2,
                                              figsize=(10,5))
# plot
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

### subplot option 2

```python
fig, ax = plt.subplots(nrows=4,ncols=4,figsize=(10,5))
# plot for each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

```python
# Days 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
heart = pd.read_csv('heart-disease.csv')
heart
```

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
# create a histogram of age 
heart['age'].plot.hist();
```

```python
heart.plot.hist(figsize =(30,30), subplots=True);
```

```python
heart.head()
```

```python
over_50 = heart[heart['age'] > 50]
over_50.head()
```

```python
# pyplot method (plotting directly from dataframe)
over_50.plot(x='age',y='chol', kind='scatter',c='target');
```

```python
#  OO method 
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);

ax.set_xlim([45, 100]);
```

```python
fig
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);
ax.set(title='Scatter Graph', xlabel='Age', ylabel='Cholestral');

# add legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
fig, ax = plt.subplots(figsize=(10,6))

bar = ax.bar(over_50['age'],height=over_50['chol']);
```

```python
# subplot  of chol, age, and thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),nrows=2,ncols=1,sharex=True)

# sub plot 1
scatter0 = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax0.set(title='Chol vs Age',ylabel='Chol');

ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(over_50['chol'].mean(),linestyle='--');

# sub plot 2
scatter1 = ax1.scatter(x=over_50['age'], y=over_50['thalach'],c=over_50['target'])
ax1.set(title='Thalach vs Age', xlabel='Age', ylabel='Thalach')

ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(over_50['thalach'].mean(),linestyle='--');

# Add title to figure
fig.suptitle('Heart Desease Analysis', fontsize=14, fontweight='bold');
```

```python
# Customizing purpose
# subplot  of chol, age, and thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),nrows=2,ncols=1,sharex=True)

# sub plot 1
scatter0 = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='winter')
ax0.set(title='Chol vs Age',ylabel='Chol');

ax0.set_xlim([50, 80])

ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(over_50['chol'].mean(),linestyle='--');

# sub plot 2
scatter1 = ax1.scatter(x=over_50['age'], y=over_50['thalach'],c=over_50['target'],cmap='winter')
ax1.set(title='Thalach vs Age', xlabel='Age', ylabel='Thalach')

ax1.set_ylim([60,200])

ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(over_50['thalach'].mean(),linestyle='--');

# Add title to figure
fig.suptitle('Heart Desease Analysis', fontsize=14, fontweight='bold');
```
